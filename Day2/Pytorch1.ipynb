#Tensor datatypes
default datatype is float 32
device can be cuda

**Note: Tensor dataype is one of the 3 big issues you will find in pytord and in deep learning
1. Tensor of not right shape
2. dtype
3. right device

```python
#Float 32 tensor
Float_32_tensor = torch.tensor([3.0, 6.0, 9.0], dtype= None
                               , device =None #what device is tensor on
                               , requires_grad=False # wether or not to use tensor to track gradeint
                               )
Float_32_tensor
```

```python
Float_32_tensor.dtype
```

```python
Float_16_tensor= Float_32_tensor.type(torch.float16)
Float_16_tensor
```
